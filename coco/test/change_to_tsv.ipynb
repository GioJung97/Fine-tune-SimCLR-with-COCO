```python
import json
import csv

def json_to_tsv(json_path, tsv_path):
    # Load JSON
    with open(json_path, "r", encoding="utf-8") as f:
        data = json.load(f)

    # If JSON is a dict, convert to list
    if isinstance(data, dict):
        data = list(data.values())

    if not isinstance(data, list) or len(data) == 0:
        raise ValueError("JSON must be a non-empty list or dict of records")

    # Collect all possible keys (robust to missing fields)
    fieldnames = sorted({key for item in data for key in item.keys()})

    # Write TSV
    with open(tsv_path, "w", encoding="utf-8", newline="") as f:
        writer = csv.DictWriter(
            f,
            fieldnames=fieldnames,
            delimiter="\t",
            extrasaction="ignore"
        )
        writer.writeheader()
        writer.writerows(data)

    print(f"Saved TSV file to: {tsv_path}")

json_to_tsv("_annotations.coco.json", "output.tsv")
```
